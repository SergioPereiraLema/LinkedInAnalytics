```python
#Importamos las librerías
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
#cargamos el fichero de conexiones
df = pd.read_csv("Connections3.csv",";")
df.head(10)
```

```python
df.describe()
```

```python
df['Connected On'] = pd.to_datetime(df['Connected On'])
```

```python
#visualizamos las conexiones por fecha
fig1 = px.line(df.groupby(by="Connected On").count().reset_index(), x="Connected On", y="First Name", labels={"First Name": "Count"}, title="Nro de conexiones")
fig1.show()
```

# Company analisis

```python
# Agrupamos información por compañía y ordenamos
```

```python
df_by_company = df.groupby(by="Company").count().reset_index().sort_values(by="First Name", ascending=False).reset_index(drop=True)
```

```python
df_by_company
```

```python
fig2 = px.bar(df_by_company[:20], x="Company", y="First Name", labels={"Company", "count"}, title="Contactos por empresa")
fig2.show()
```

```python
fig3 = px.treemap(df_by_company[:100], path=["Company"], values="First Name", labels={"Company", "Count"})
fig3.show()
```

### Análisis por puesto

```python
df_by_position = df.groupby(by="Position").count().reset_index().sort_values(by="First Name", ascending=False).reset_index(drop=True)
df_by_position
```

### Creamos visualizaciones para el análisis por puesto

```python
fig4 = px.bar(df_by_position[:20], x="Position", y="First Name", labels={"Position", "Count"}, title="Contactos por puesto")
fig4.show()
```

```python
fig5 = px.treemap(df_by_position[:100], path=["Position"], values="First Name", labels={"Position", "Count"}, title="Mapa de contactos por posición")
fig5.show()
```

```python
df_by_Name = df.groupby(by="First Name").count().sort_values(by="Company", ascending=False).reset_index()
df_by_Name
```

```python
fig6 = px.treemap(df_by_Name[:100], path=["First Name"], values="Company", labels={"Name", "Count"}, title="Contactos por nombre")
fig6.show()
```
